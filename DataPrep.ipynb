```python
# Import Section

import numpy as np
import pandas as pd 
from IPython.display import display
```

```python
#Displaying the Data (Raw Format)

D1 = pd.read_excel("C:\KD\Python Learning\Projects\YoutubeStat\Excel Files\YT_GlobalS.xlsx")
display(D1)
```

```python
#Checking Data Info

D1.dtypes
```

```python
#Checking Duplicates

(D1[D1.duplicated()]).count()

#Output : 0 Duplicates
```

```python
#Checking Null Values
(D1[D1.isnull()]).count()
```

```python
#checking Legit Youtubers (not # channels)

D1[D1['video views'].isin([0])]
```

```python
#removing the channels

D1 = D1.drop(D1[D1['video views'] == 0].index)
display(D1.head())

#checking again

D1[D1['video views'].isin([0])]
```

```python
#checking whole dataset for value 0

(D1[D1.isin([0])]).count()

# exception made columns values can be 0.
```

```python
#Checking for other anamolies

D1 = D1.replace('nan', pd.NA)  # Replace 'nan' with pandas' NA for numeric columns
numeric_columns = ['rank', 'subscribers', 'video views', 'video_views_for_the_last_30_days',
                   'lowest_monthly_earnings', 'highest_monthly_earnings', 'lowest_yearly_earnings',
                   'highest_yearly_earnings', 'subscribers_for_last_30_days', 'created_year', 'Unemployment rate']
for col in numeric_columns:
    D1[col] = pd.to_numeric(D1[col], errors='coerce')
    
D1 = D1.dropna()

#cleared and replace all incorrect strings and null values

(D1[D1.isin([0])]).count()
```
